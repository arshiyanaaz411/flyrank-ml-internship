## The Question: Decision, Action, Cost of a Wrong Call

**Research Question:**
Which content pages have the highest opportunity to improve
CTR and engagement, given their current search impressions
and average position?

**Unit of Analysis:**
One row = one content page (content_hash_id)

**Output:**
A ranked list of pages scored by CTR opportunity —
highest priority pages appear first.

**Decision:**
Which pages should a content reviewer look at first?

**Action:**
Update page title, meta description, or content structure
to improve click-through rate from search results.

**Cost of a Wrong Call:**
- False positive: reviewer wastes time on a page that
  didn't need fixing
- False negative: a high-opportunity page gets ignored,
  client loses potential traffic
- Missing high-impression, low-CTR pages = direct
  revenue loss for FlyRank clients## 2. The question: decision, action, cost of a wrong call

## Quick Look at the Data: Supporting Numbers

### What the data shows:

- **Total pages analysed:** 30,000

  get impressions but very few clicks
- **High impression, Low CTR pages: 14,245 (47.5% of all pages)**

- **Median position:** 10.8 — many pages are on page 1 of Google
